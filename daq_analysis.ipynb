```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog
import os

root = tk.Tk()
root.withdraw()
file = filedialog.askopenfile()
df = pd.read_csv(file.name)
df.set_index('time_stamp', inplace=True)
num_signals = len(df.columns)
df
```

```python
df.ffill()
```

```python
import matplotlib.pyplot as plt

# Create a figure and array of subplots with shared x-axis
fig, axes = plt.subplots(nrows=num_signals, ncols=1, figsize=(10, num_signals * 3), sharex=True)

# Loop through each sensor and create a subplot
for i, column in enumerate(df.columns):
    axes[i].plot(df.index, df[column], label=column)
    axes[i].set_title(column)
    axes[i].set_ylabel('Value')
    axes[i].legend(loc='upper right')

# Set a common x-label
plt.xlabel('Timestamp')

# Improve layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(30, 6))

# Plot the first data series
color = 'tab:red'
ax1.set_xlabel('Time (millis)')
ax1.set_ylabel('DC bus voltage', color=color)
ax1.plot(df.index, df['packVoltage'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for the second data series
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('DC Bus Current', color=color)
ax2.plot(df.index, df['INV_DC_Bus_Current'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create a third y-axis
ax3 = ax1.twinx()
# Offset the third axis
ax3.spines['right'].set_position(('outward', 60))
color = 'tab:green'
ax3.set_ylabel('Motor speed (rpm)', color=color)
ax3.plot(df.index, df['motorSpeed'], color=color)
ax3.tick_params(axis='y', labelcolor=color)
ax1.set_xlim(0, 200000)
# Show the plot
plt.title('Triple Axis Time Series Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(30, 6))

# Plot the first data series
color = 'tab:red'
ax1.set_xlabel('Time (millis)')
ax1.set_ylabel('torqueRequest', color=color)
ax1.plot(df.index, df['torqueRequest'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for the second data series
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('inv_torque_request', color=color)
ax2.plot(df.index, df['inv_torque_request'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Create a third y-axis
ax3 = ax1.twinx()
# Offset the third axis
ax3.spines['right'].set_position(('outward', 60))
color = 'tab:green'
ax3.set_ylabel('inv_torque feedback', color=color)
ax3.plot(df.index, df['inv_torque feedback'], color=color)
ax3.tick_params(axis='y', labelcolor=color)

ax1.set_xlim(0, 200000)

# Show the plot
plt.title('torque data')
plt.show()
```
